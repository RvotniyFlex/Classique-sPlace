**Задание 10**

Максим решил проверить гипотезу о том, что глицин улучшает показатели учеников на проверочных. Для этого он отобрал две группы учеников: одной он в течение месяца на своих занятиях выдавал глицин, а другой - не выдавал. По результатом его замеров групп без глицина и с глицином вышли соответствующие результаты: 

*x_бгл = [7, 8, 6, 5, 7, 6, 7, 8, 9, 10]*

*x_гл = [8, 5, 7, 9, 10, 6, 7, 6, 9, 8]*

Проверьте t-тест для двух выборок с одинаковыми дисперсиями:

$$
\begin{aligned}
&H_0: \overline{x_1} = \overline{x_2}\\
&H_a: \overline{x_1} < \overline{x_2}
\end{aligned}
$$

Выберите все верные утверждения:

* Нулевая гипотеза не отвергается на уровне значимости 1% **(верно)**
* Альтернативная гипотеза не отвергается на уровне значимости 1%
* Нулевая гипотеза не отвергается на уровне значимости 5% **(верно)**
* На уровне значимости 5% можно утверждать, что ученики с глицином лучше пишут тесты 

```python
from scipy.stats import ttest_ind
```

```python
x_1 = [7, 8, 6, 5, 7, 6, 7, 8, 9, 10]
x_2 = [8, 5, 7, 9, 10, 6, 7, 6, 9, 8]
```

```python
ttest_ind(x_1, x_2, alternative='less')
```
