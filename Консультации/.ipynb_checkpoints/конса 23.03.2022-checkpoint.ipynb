```python
pattern = 'https://www.russianfood.com/' #создадим шаблон, чтобы формировать ссылку
```

```python
print(pattern + tree.find('a', {'name': 'el60031'}).get('href'))
```

```python
cards = tree.find_all('div', {'class': 'recipe_l in_seen v2'})
```

```python
infa = {'name': [], #словарь, куда будем заносить результаты парсинга
       'url': []
       }
```

```python
cards[0].find('a', {'name': 'el60031'}).get('href')
```

```python
for card in cards:
    infa['name'].append(card.find('div', {'class': 'title'}).a.text)
    infa['url'].append(pattern + card.find('div', {'class': 'title'}).a.get('href'))
```

```python
card.find('div', {'class': 'title'}).a.get('href')
```

```python
infa['description'] = [] #новый ключ, где будем хранить описание 
```

```python
for link in infa['url']:
    response = requests.get(link, headers={'User-Agent': UserAgent().chrome}) ###
    tree = bs(response.content, 'html.parser')
    try:
        infa['description'].append(tree.find('td', {'class': 'padding_l padding_r'}).find('div', {'class': None}).p.text)
    except:
        infa['description'].append('-')
        pass
    time.sleep(2)
```
