```python
import pandas as pd
```

```python
df = pd.read_csv(r'DFs/cian_2022.csv', index_col=0 )
```

```python
df
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df = df.drop(['region', 'metro', 'adress', 'type_of_way', 'restroom', 'price_per_square', 'building_age', 'rooms_in_flat', 'window_view', 'district', 'to_metro', 'apartment_layout', 'remont', 'age_of_construction', 'rooms_to_rent', 'height', 'share_in_flat'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15,9))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', ax=ax);
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['price']
X = df.drop(['price'], axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
np.random.seed(2)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print ("Train: {}".format(mean_squared_error(y_train, lr.predict(X_train))))
print ("Test: {}".format(mean_squared_error(y_test, lr.predict(X_test))))
```

```python
for col, coef in zip(X.columns, lr.coef_):
    print('{}: {:.2f}'.format(col, coef))
```

```python
w = np.sum(lr.coef_)
```

```python
for col, coef in zip(X.columns, lr.coef_):
    print('{}: {:.2f}%'.format(col, coef/w*100))
```
