```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def abm_path_generator(n_paths, n_steps, T, mu, sigma, X_0):

    # Create 
    Z = np.random.normal(0.0, 1.0, [n_paths, n_steps])
    X = np.zeros([n_paths, n_steps + 1])
    time = np.zeros(n_steps + 1)
    
    X[:, 0] = X_0
    dt = T / float(n_steps)

    for k in range(0, n_steps):
        X[:, k+1] = X[:, k] + mu * dt + sigma * np.sqrt(dt) * Z[:, k]
        time[k+1] = time[k] + dt
    
    paths = {'X': X, 'time': time}

    return paths
```

```python
def main_calculation():
    n_paths = 25
    n_steps = 10000
    T = 1
    mu = 100
    sigma = 40
    X_0 = 0

    paths = abm_path_generator(n_paths, n_steps, T, mu, sigma, X_0)
    ABM = paths['X']
    time_grid = paths['time']

    plt.figure(figsize = (10, 6))
    plt.plot(time_grid, np.transpose(ABM))
    plt.grid()
    plt.xlabel('t')
    plt.ylabel('X(t)')
```

```python
main_calculation()
```
